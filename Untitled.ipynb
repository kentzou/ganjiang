```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['Microsoft YaHei']
```

```python
df_1994 = pd.read_excel(io="data/1994,2000,2014.xlsx",sheet_name=0).drop(axis=0,index=0)
df_2000 = pd.read_excel(io="data/1994,2000,2014.xlsx",sheet_name=1).drop(axis=0,index=0)
df_2014 = pd.read_excel(io="data/1994,2000,2014.xlsx",sheet_name=2).drop(axis=0,index=0)

df1 = df_1994.sort_values(by = "南昌",ascending=0)
df2 = df_2000.sort_values(by = "南昌",ascending=0)
df3 = df_2014.sort_values(by = "南昌",ascending=0)

df1 = df_1994[["日期","外洲","南昌","吴城","康山"]]
df1.columns = ["时间","外洲","南昌","吴城","康山"]
df1= df1.append(df_2000[["时间","外洲","南昌","吴城","康山"]])
# df1= df1.append(df_2014[["时间","外洲","南昌","吴城","康山"]])
df1 = df1.sort_values(by = "南昌",ascending=1)
df_1994.head()
```

```python
def get_ins(col,df):
    nc = df[col].values.tolist()
    del(nc[0])
    nc.append(0)
    df["nc"] =nc
    df["T"] = df["nc"]-df1[col]
    data= df["T"].values.tolist()
    df.drop(["nc","T"],axis = 1,inplace = True)
    return data

```

```python
df1["南昌增量"]=get_ins("南昌",df1)
df1["康山增量"]=get_ins("康山",df1)
df1["吴城增量"]=get_ins("吴城",df1)
df1 =df1.reset_index().drop("index",axis=1)
df1.drop(axis=0,index=df1.shape[0]-1,inplace=True)
df1.head()
```

```python
df2 = df1.sort_values(by="外洲",ascending=1)
df2.head()
```

```python
#滑动平均
window=10
df1["吴城平均"] = df1["吴城"].rolling(window=window).mean()
df1["康山平均"] = df1["康山"].rolling(window=window).mean()
df1 = df1.dropna()
df1.head()
```

```python
def m_plt(y1,y2,save=False):
    #计算拟合
    x = [i for i in range(df1.shape[0])]
    m = np.polyfit(x, df1[y2].values.tolist(), 3)
    y = [m[0] * np.power(i,3) + m[1] * np.power(i,2) + m[2]* np.power(i,1)+ m[3] for i in x]
    
    plt.figure(figsize=(15, 8))
    plt.title(y1+y2+"水位变化")
    plt.ylabel("水位")
    plt.plot(df1[y1],label = y1+"水位")
    plt.plot(df1[y2],label = y2+"水位")
    #plt.plot(y,label = y2+"拟合")
    plt.plot(df1[y2+"平均"],label = y2+"平均")
    plt.grid()
    plt.legend()
    if save:
        plt.savefig(y1+y2+".png")
```

```python
m_plt("南昌","吴城")
```

```python
m_plt("南昌","康山")
```

```python
plt.figure(figsize=(15, 8))
plt.title("南昌水位与流量关系")
plt.xlabel("南昌水位")
plt.ylabel("南昌流量")
plt.plot(df1["南昌"],df1["外洲"])
plt.grid()
plt.legend()
```
